```python
connection = psycopg2.connect(dbname='films')

try:
    with connection:
        with connection.cursor(cursor_factory=extras.DictCursor) as cursor:
            cursor.execute('SELECT * FROM directors')
            rows = cursor.fetchall()
finally:
    connection.close()

for row in rows:
    print(row['name'])
```

```python
import datetime

# Create a datetime object
now = datetime.datetime.now() 
print(f"Original datetime object: {now}")

# Extract the date part
date_only = now.date()
print(f"Date part: {date_only}")

# You can also create a specific datetime object and extract its date
specific_datetime = datetime.datetime(2025, 10, 18, 10, 30, 0)
print(f"Specific datetime object: {specific_datetime}")

specific_date_only = specific_datetime.date()
print(f"Date part of specific datetime: {specific_date_only}")
```

```python
from contextlib import contextmanager
```

```python
@contextmanager
def customer_db_connection(db_name):
    connection = psycopg2.connect(dbname = db_name)
    try:
        with connection:
            yield connection
    finally:
        connection.close()
        
     
```

```python
with customer_db_connection(db_name = 'expenses') as connection:
    with connection.cursor() as cursor:
        cursor.execute("INSERT INTO expenses (amount, memo) VALUES (3, 'Do not care')")
        #rows = cursor.fetchall()
```

```python
with customer_db_connection(db_name = 'expenses') as connection:
    with connection.cursor() as cursor:
        cursor.execute("SELECT * FROM expenses")
        rows = cursor.fetchall()
```

```python
for row in rows:
    print(row)
```

```python
@contextmanager
def outer():
    print("Enter Outer")
    try:
        yield inner()
    finally:
        print("Exit Outer")

@contextmanager
def inner():
    print("Enter Inner")
    try:
        yield
    finally:
        print("Exit Inner")

with outer() as i:
    with i:
        print("Hi")
```
