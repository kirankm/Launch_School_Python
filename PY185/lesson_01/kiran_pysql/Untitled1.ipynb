```python
import re 

def get_param_dict(request_line):
    param_dict = {}
    matches = re.findall(r'[\w]+=.', request_line)
    for match in matches:
        key, val = match.split("=")
        param_dict[key] = int(val)
    return param_dict

import socket
import random

server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

server_socket.bind(('localhost', 3006))
server_socket.listen()

while True:
    client_socket, addr = server_socket.accept()
    print(f"Connection from {addr}")

    request = client_socket.recv(1024).decode()
    if (not request) or ('favicon.ico' in request):
        client_socket.close()
        continue

    request_line = request.splitlines()[0]
    param_dict = get_param_dict(request_line)

    roll_arr = [random.randint(1, param_dict['sides']) for roll in range(param_dict['rolls'])]
    response_body = f"{request_line}\n"
    for index, val in enumerate(roll_arr):
        response_body += f'Roll {index + 1}:{val}\n'

    response = ("HTTP/1.1 200 OK\r\n"
            "Content-Type: text/plain\r\n"
            f"Content-Length: {len(response_body)}\r\n"
            "\r\n"
            f"{response_body}\n")

    client_socket.sendall(response.encode())
    client_socket.close()
```
